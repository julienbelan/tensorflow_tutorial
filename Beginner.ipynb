# NEURAL NETWORK

At the core of neural network is the perceptron. Perceptrons are simply a specific family of functions that mimic a trigger response from a nerve cell. Its a simple linear function:

activation = weight x input+bias

To which we add a weird function that triggers only if a certain threshold is met, in other words, the function goes to 0 if the activation is less or equal to the threshold and goes to 1 else.

output = 1 if activation > threshold & 0 if activation <= threshold

```python
# A perceptron function, in the python language, def is placed before a name to introduce a function, arguments are
# placed between parantheses. I.e: def function_name(arguments 1, arguments 2):
def perceptron(input, weight, bias, threshold):
    # Here the code representation of the 
    if weight*input+bias > threshold:
        return 1
    else:
        return 0
```

```python
Image.open('images/perceptron.png')
```

Seems like hard coding? The real magic happens when you connect multiple perceptron together. To handle multiple outputs, we sum them respectively to their weights.

```python
# Calling a function from another function is a common way in python to avoid reinventing the wheel
def multi_input_perceptron(inputs, weights, bias, threshold):
    inputs = sum(inputs.dot(weights))+bias
    return perceptron(inputs, threshold)
```
